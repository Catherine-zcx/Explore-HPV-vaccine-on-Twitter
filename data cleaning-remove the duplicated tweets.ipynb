```python
import pandas as pd
```

```python
df1 = pd.read_csv('cervarix0408.csv') #open the csv
df2 = pd.read_csv("cervarixvaccine0408.csv")
df3 = pd.read_csv('gardasil0408.csv')
df4 = pd.read_csv('hpvvaccine0408.csv')
df5 = pd.read_csv('cervarix0417.csv') #open the csv
df6 = pd.read_csv("cervarixvaccine0417.csv")
df7 = pd.read_csv('gardasil0417.csv')
df8 = pd.read_csv('hpvvaccine0417.csv')
df9 = pd.read_csv('cervarix0427.csv') #open the csv
df10 = pd.read_csv("cervarixvaccine0427.csv")
df11 = pd.read_csv('gardasil0427.csv')
df12 = pd.read_csv('hpvvaccine0427.csv')
df13 = pd.read_csv('cervarix0503.csv') #open the csv
df14 = pd.read_csv("cervarixvaccine0503.csv")
df15 = pd.read_csv('gardasil0503.csv')
df16 = pd.read_csv('hpvvaccine0503.csv')
df17 = pd.read_csv('cervarix0509.csv') #open the csv
df18 = pd.read_csv("cervarixvaccine0509.csv")
df19 = pd.read_csv('gardasil0509.csv')
df20 = pd.read_csv('hpvvaccine0509.csv')


df = df1.append([df2, df3,df4,df5,df6,df7,df8,df9,df10,df11,df12,df13,df14,df15,df16])
df.head()
```

```python
print([column for column in df])
```

```python
df = df[['screen_name',"text","is_retweet","favorite_count",
         "retweet_count","verified",
         "followers_count"]]
df["text"]= df["text"].astype(str)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df = df.sort_values(by = 'is_retweet',axis = 0,ascending = True)
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
df = df.drop_duplicates(subset='text', keep='first', inplace=False)
df
```

```python
df = df.sort_values(by = ['retweet_count','verified'],axis = 0,ascending = [False,True])
df
```
